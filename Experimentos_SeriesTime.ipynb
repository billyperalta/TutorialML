```python
#code for extracting time series of dataset over a random sampling of dataset
import matplotlib.pyplot as plt
from random import randint
from matplotlib.backends.backend_pdf import PdfPages
from tsfresh import extract_features

unique=sales_df[10].drop_duplicates()
unique = unique.reset_index(drop=True)
numrows=len(unique.index) #numero de vehiculos
print(numrows)
#Generate a 10 random samples of time series
#we add a code for extracting feature over time serie
pp = PdfPages('foo1.pdf')
for x in range(0, 10):
    #Generate random number
    #f[x] = plt.figure()    
    indice=randint(0, numrows)
    #print(unique[indice])
    df = sales_df[4].loc[sales_df[10] == unique[indice]].copy()
    #fig = plt.figure(figsize=(8.27, 11.69), dpi=100)
    #1. print time serie in PDF file; only for checking the values
    df1 = sales_df.loc[sales_df[10] == unique[indice]].copy()
    fig = plt.figure()
    df.plot(subplots=True, sharex=True, figsize=(5,5))
    plt.show()  
    #plt.close()    
    pp.savefig(fig)
    #f.close()
    
    #2. only checking extraction of features
    extracted_features = extract_features(df1, column_id=10, column_sort=0)
    
pp.close()
```
